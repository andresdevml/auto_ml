```python
from pycaret.classification import *

import numpy as np

import pandas as pd

from sklearn.metrics import roc_auc_score

from sklearn.metrics import accuracy_score

from sklearn.metrics import recall_score

from google.colab import drive
```

# **Creamos la data de entrada**

```python
# conectamos al drive

drive.mount('/content/drive')

# Dfinimos el directorio

dir_data='./drive/MyDrive/PyCaret_Santander/data'

# definimos la data

mask_length=pd.read_csv(
                filepath_or_buffer=dir_data+'/train.csv'
                        ).shape[0]

mask_index=np.random.rand(mask_length)

pred_index=mask_index<=0.05

df_pred=pd.read_csv(filepath_or_buffer=dir_data+'/train.csv')[pred_index]

y_test=df_pred['TARGET']

df_pred.drop(labels=['TARGET'],axis = 1,inplace = True)


```

# **Importamos el modelo**

```python
model=load_model(model_name=dir_data+'/model_pipeline')
```

# **Prediciones y verificacion**

```python
y_answer=predict_model(
                  estimator=model,
                  raw_score = True,
                  data = df_pred
                                                          )

metric_auc=roc_auc_score(y_test,y_answer['prediction_score_1'])

metric_acc=accuracy_score(y_test,y_answer['prediction_label'])

metric_rec=recall_score(y_test,y_answer['prediction_label'])


print(f'Para el Stack_model:\n\nauc={np.round(metric_auc,3)}\n\naccuracy={np.round(metric_acc,3)}\n\nrecall={np.round(metric_rec,3)}')
```
