```python
from pycaret.classification import *

import numpy as np

import pandas as pd

from sklearn.metrics import roc_auc_score

from sklearn.metrics import accuracy_score

from sklearn.metrics import recall_score

from google.colab import drive
```

```python
# picamos la data para hacer la verificacion final

mask_length=pd.read_csv(
                filepath_or_buffer=dir_data+'/train.csv'
                        ).shape[0]

mask_index=np.random.rand(mask_length)

train_index=mask_index<=0.9

test_index=mask_index >0.9

# generamos los datasets

df_train=pd.read_csv(filepath_or_buffer=dir_data+'/train.csv')[train_index]


df_test=pd.read_csv(filepath_or_buffer=dir_data+'/train.csv')[test_index]

```

# **Definimos la tuberia de preprocesado y entrenamiento**

```python
# definimos el setup

clf= setup(

          # preprocesing
           data = df_train,
           ignore_features = ['ID'],
           target = 'TARGET',
           fix_imbalance=True,

          # Scaler

          normalize=True,
          normalize_method='zscore',

          # FEATURE SELECTION

          feature_selection=True,

          feature_selection_method='classic',

          feature_selection_estimator='lightgbm',

          n_features_to_select=0.5,

          remove_multicollinearity=True,

          multicollinearity_threshold=0.9,

          # SETUP PARAMETERS


          use_gpu=True,

          train_size=0.7,

          fold_strategy='stratifiedkfold',

          fold=5,

          fold_shuffle=True

           )
```

```python
# definimos el modelo


model=compare_models(sort='AUC',n_select =1,
                       include=['lightgbm','gbc','rf','dt','xgboost'])

```

```python
# finalizamos el modelo



finalize_model(model)
```

```python
# probamos el modelo


y_test=df_test['TARGET']

y_pred=predict_model(
                  estimator=model,
                  raw_score = True,
                  data = df_test.drop(
                                         labels=['TARGET'],
                                          axis = 1,
                                          inplace = False)
                                                          )

metric_auc=roc_auc_score(y_test,y_pred['prediction_score_1'])

metric_acc=accuracy_score(y_test,y_pred['prediction_label'])

metric_rec=recall_score(y_test,y_pred['prediction_label'])


print(f'Para el Stack_model:\n\nauc={np.round(metric_auc,3)}\n\naccuracy={np.round(metric_acc,3)}\n\nrecall={np.round(metric_rec,3)}')
```

```python
# nos llevamos el modelo


save_model(model=model, model_name=dir_data+'/model_pipeline')
```
